*Inspecting the data set and treating missing values*

```python
#Inspect missing values in data set
print(df.isnull().values.sum())

#Replace the ' 's with NaN
df=df.replace(" ",np.NaN)

#Impute the missing values with mean imputation
df=df.fillna(df.mean())

#Count the number of NaNs in the dataset to verify
print(df.isnull().values.sum())
```

*Since there are no missing values, we can now finally begin the EDA*

### *Plotting a histogram*

```python
#seaborn histogram
sns.histplot(df['Age'], kde=False, bins=10, binwidth=2, color="pink", edgecolor="black")

#Add the label in the graph
plt.title("Age Count")
plt.xlabel("Age") 
plt.ylabel("Count") 
```

* The above data illustrates the age difference among employees

```python
sns.displot(data=df,x='Age',kind='kde',color='pink')

plt.title("Age Count")
plt.xlabel("Age")
plt.ylabel("Count")
```

```python
sns.histplot(df['ExperienceInCurrentDomain'], kde=True, bins=10, binwidth=2, color="purple", edgecolor="black")

plt.title("Experience Level")
plt.xlabel("Experience")
plt.ylabel("Level")
```

```python
sns.displot(data=df, x='Age',y='Gender')
plt.title("Age vs Gender")
plt.xlabel("Age")
plt.ylabel("Gender")
```

* The above histogram shows the relation between age vs gender
* Equal distribution can be seen between both genders

### *Plotting a bar graph*

```python
sns.barplot(data=df, x="Gender", y="Age")

plt.xlabel("Gender")
plt.ylabel("Age")
plt.title("Gender vs Age")
```

### *Plotting a lineplot*

```python
sns.lineplot(data=df, x="Education", y="ExperienceInCurrentDomain", color="purple")
plt.title("Education vs Experience")
plt.xlabel("Education")
plt.ylabel("Experience")
plt.show()
```

* The above graph shows the relation between education and experience.


```python
sns.lineplot(data=df, x="Age", y="Gender", color="green")
plt.title("Age vs Gender")
plt.xlabel("Age")
plt.ylabel("Gender")
plt.show()
```

### *Plotting a scatterplot*

```python
sns.scatterplot(data=df, x="ExperienceInCurrentDomain",y="Education", color="maroon")
plt.title("Education vs Experience")
plt.xlabel("Experience")
plt.ylabel("Education")
```

```python
sns.scatterplot(data=df, x="Gender",y="Age", color="maroon")
plt.title("Age vs Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
```

* Equal distribution can be found between both genders

### *Plotting a boxplot*

```python
sns.boxplot(data=df, x="Age", y="Gender")
plt.xlabel("Age")
plt.ylabel("Gender")
plt.title("Age vs Gender")
```

* Both of the boxplot seem to be right skewed.
* The age ranges from 20 to 45
* The median of both the gender seem to be almost 28
* The above data is dispersed from 20 to 45
* No outliar is found in the data


```python
sns.boxplot(data=df, y="Education", x="ExperienceInCurrentDomain")
plt.title("Experience vs Education")
plt.xlabel("Experience")
plt.ylabel("Education")
```

* The above data illustrates the relation between education and experience
* The median of all of the three both plot is 3
* None of them are skewed
* The experience of the Bachelors as well as Masters ranges from 0 to 7 years but that of PHD only ranges from 0 to 5 years
* It shows that the employee having a PHD degree have less experince.
